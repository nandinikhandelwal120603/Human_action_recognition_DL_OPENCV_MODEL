```python
#creating csv 
import os
import csv

# Update the list of classes if needed
classes = [
    "BlowDryHair",
    "BrushingTeeth",
    "CuttingInKitchen",
    "Haircut",
    "MoppingFloor",
    "ShavingBeard",
    "Typing",
    "WritingOnBoard"
]

# Update the paths accordingly
folder_root = r"C:\Users\khand\OneDrive\Desktop\CVR"
data_root= r"C:\Users\khand\OneDrive\Desktop\CVR\dataset"

# Update the path where you want to save the CSV file
csv_path = os.path.join(folder_root, 'dataset.csv')

with open(csv_path, 'w', newline='') as file:
    writer = csv.writer(file)
    for c in classes:
        path = os.path.join(data_root, c + "/")
        for i in os.listdir(path):
            video_path = os.path.join(path, i)
            class_index = classes.index(c)
            writer.writerow([class_index, video_path])

# Check if the CSV file was created
if os.path.exists(csv_path):
    print(f"CSV file '{csv_path}' successfully created.")
else:
    print(f"CSV file '{csv_path}' creation failed.")
```
