```python
import cv2
import numpy as np

IMG_SIZE = (224, 224)  # Define your desired image size here

def crop_center_square(frame):
    y, x = frame.shape[0:2]
    min_dim = min(y, x)
    start_x = (x // 2) - (min_dim // 2)
    start_y = (y // 2) - (min_dim // 2)
    return frame[start_y: start_y + min_dim, start_x: start_x + min_dim]

def load_video(path, max_frames=0, resize=IMG_SIZE):
    cap = cv2.VideoCapture(path)
    frames = []
    j = 0
    try:
        while True:
            ret, frame = cap.read()
            if not ret:
                break
            frame = crop_center_square(frame)
            frame = cv2.resize(frame, resize)
            frame = frame[:, :, [2, 1, 0]]
            frames.append(frame)
            j += 1
            if max_frames > 0 and j >= max_frames:
                break
    finally:
        cap.release()
    return np.array(frames)

# Example usage:
video_path = "path_to_your_video.avi"
frames = load_video(video_path, max_frames=10)  # Load up to 10 frames from the video
```

```python
import keras
from keras.preprocessing.image import ImageDataGenerator
from keras.layers import Input, Conv2D, MaxPooling2D, Flatten, Dense
from keras.models import Model

# Define the path to your training data folder
train_data_dir = r"C:\Users\khand\OneDrive\Desktop\CVR\train"

# Define the image size and batch size
IMG_SIZE = 224
BATCH_SIZE = 32
EPOCHS = 10

# Create an ImageDataGenerator for data augmentation and preprocessing
train_datagen = ImageDataGenerator(
    rescale=1.0/255.0,
    rotation_range=20,
    width_shift_range=0.2,
    height_shift_range=0.2,
    shear_range=0.2,
    zoom_range=0.2,
    horizontal_flip=True,
    fill_mode='nearest'
)

# Create a data generator for your training data
train_generator = train_datagen.flow_from_directory(
    train_data_dir,
    target_size=(IMG_SIZE, IMG_SIZE),
    batch_size=BATCH_SIZE,
    class_mode='categorical',
    shuffle=True
)

# Define your sequence model
# Replace this with your actual model architecture
input_layer = Input(shape=(IMG_SIZE, IMG_SIZE, 3))  # Adjust input shape as needed
x = Conv2D(64, (3, 3), activation='relu')(input_layer)
x = MaxPooling2D(pool_size=(2, 2))(x)
x = Flatten()(x)
x = Dense(128, activation='relu')(x)
output_layer = Dense(7, activation='softmax')(x)  # Assuming 7 classes
sequence_model = Model(inputs=input_layer, outputs=output_layer)

# Compile your model
sequence_model.compile(
    optimizer=keras.optimizers.Adam(learning_rate=1e-5),  # Use learning_rate instead of lr
    loss='categorical_crossentropy',
    metrics=['accuracy']
)

# Train your model using the generator
history = sequence_model.fit(
    train_generator,
    epochs=EPOCHS,
    steps_per_epoch=len(train_generator),
    verbose=1
)


# Save your trained model
sequence_model.save('trained_sequence_model.h5')
```
